# 这里我对印度大陆给它挑出来

```python
lat_extent  =  [0,23]
lon_extent  =  [65,90]


new_bathy  =  np.empty([0],dtype = float)
new_lat    =  new_bathy.copy()
new_lon    =  new_bathy.copy()
import math

for i in range(0,len(f1.grid_imask)):
    if (f1.grid_center_lat.data[i] > math.radians(lat_extent[0]) and f1.grid_center_lat.data[i] < math.radians(lat_extent[1])) and (f1.grid_center_lon.data[i] > math.radians(lon_extent[0]) and f1.grid_center_lon.data[i] < math.radians(lon_extent[1])):
        new_bathy = np.append(new_bathy,f1.grid_imask[i])
        new_lat   = np.append(new_lat,f1.grid_center_lat.data[i])
        new_lon   = np.append(new_lon,f1.grid_center_lon.data[i])
```
