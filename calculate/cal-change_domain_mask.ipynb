## 本代码修改domain文件
## 更改海陆分布

```python
import xarray as xr
import numpy as np
path = "/home/sun/data/domain/"
f1   =  xr.open_dataset(path+"domain.lnd.fv0.9x1.25_gx1v7.151020.nc")
f2   =  xr.open_dataset(path+"domain.ocn.gx1v7.151008.nc")
f3   =  xr.open_dataset(path+"domain.ocn.1x1.111007.nc")

mask_zero = f1.mask.data[60,1]
frac_zero = f1.frac.data[60,1]
mask_one  = f2.mask.data[60,1]
frac_one  = f2.frac.data[60,1]
```

**这里针对mask和frac变量进行修改**

```python
for i in range(0,192):
    for j in range(0,288):
        if 70<=f1.xc.data[i,j]<=90:
            if 0<=f1.yc.data[i,j]<=20:
                f1.mask.data[i,j] = mask_zero
                f1.frac.data[i,j] = frac_zero
            else:
                continue
        else:
            continue
for i in range(0,384):
    for j in range(0,320):
        if 70<=f2.xc.data[i,j]<=90:
            if 0<=f2.yc.data[i,j]<=20:
                f2.mask.data[i,j] = mask_one
                f2.frac.data[i,j] = frac_one
            else:
                continue
        else:
            continue
```

```python
f1.to_netcdf("/home/sun/data/domain/domain.lnd.changed.nc")
f2.to_netcdf(path+"domain.ocn.gx1v7.changed.nc")
```
