```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

f = xr.open_dataset('/Users/sunweihao/data/model_data/spinup/spinup_sst_data.nc')
f
```

```python
fig    =  plt.figure()
ax     =  fig.add_subplot(111)

ax.set_xlim((0, 110))
ax.set_ylim((14, 16))

ax.plot(f.time_control.data,f.control_global_sst,color='black',lw=2.5,label='CONTROL')
ax.plot(f.time_inch.data,f.inch_global_sst,color='blue',lw=2,label='INCH')
ax.plot(f.time_indian.data,f.indian_global_sst,color='peru',lw=2,label='INDIAN')
ax.plot(f.time_maritime.data,f.maritime_global_sst,color='red',lw=2,label='MARITIME')

ax.legend(loc='upper left')

ax.set_title("Global-Mean SST")
```
