```python
'''
本代码使用模式数据来看爆发当日的纬向偏差
'''
import os
import numpy as np
import xarray as xr
import json

def query_bobsm_date(year):
    '''本函数可以根据输入的year年份返回当年的爆发日期,year type is str'''
    with open("/home/sun/data/onsetdate.json",'r') as load_f:
        ff  =  json.load(load_f)

    return int(ff[year])

def cal_zonal_t_gradient(year,lat_slice,lon_slice1,lon_slice2):
    '''year is str'''
    path1  =  "/home/sun/mydown/swh_era5/land/daily/"

    # 获取季风爆发日期
    d0  =  query_bobsm_date(year)

    # 获取爆发当日的文件
    f1  =  np.average(xr.open_dataset(path1 + year+"_2m_temperature_land_monthly.nc").sel(lat=lat_slice,lon=lon_slice1).isel(time=[d0-1]).t2m.data)
    f2  =  np.average(xr.open_dataset(path1 + year+"_2m_temperature_land_monthly.nc").sel(lat=lat_slice,lon=lon_slice2).isel(time=[d0-1]).t2m.data)

    print(f1-f2)


def main():
    cal_zonal_t_gradient("1997",lat_slice=slice(15,20),lon_slice1=slice(75,80),lon_slice2=slice(90,92))

if __name__ == '__main__':
    main()
```
