```python
'''设置绘图区域'''
lonmin,lonmax,latmin,latmax  =  30,120,-10,30
extent  =  [lonmin,lonmax,latmin,latmax]

proj    =  ccrs.PlateCarree()
fig     =  plt.figure(figsize=(36,10))
spec5 = fig.add_gridspec(nrows=3,ncols=5)

#绘制长方形
verts = [
   (90., 8.),  # left, bottom
   (110.,8.),  # left, top
   (110.,21.),  # right, top
   (90., 21.),  # right, bottom
   (0., 0.),  # ignored
]

codes = [
    Path.MOVETO,
    Path.LINETO,
    Path.LINETO,
    Path.LINETO,
    Path.CLOSEPOLY,
]

path     =  Path(verts,codes)
dates1   =  []
for dd in range(0,15):
    dates1.append("4-"+str(dd+1))
props = dict(boxstyle='Round4', facecolor='wheat', alpha=1)
```

```python
j  =  0
start  =  105
for row in range(3):
    for col in range(5):
        ax = fig.add_subplot(spec5[row,col],projection=proj)
        ax.coastlines(resolution='110m',lw=1)
        # 设置经纬度刻度.
        set_map_ticks(ax, dx=10, dy=10, nx=1, ny=1, labelsize='small')
        ax.set_extent(extent, crs=proj)
    
        ax.plot([30,120],[0,0],color='k')
        patch = patches.PathPatch(path, facecolor='grey', lw=2,alpha=0.4)
        ax.add_patch(patch)
        ax.streamplot(f0.lon.data, f0.lat.data, 
                      f0.U.data[start+j,3,:], f1.V.data[start+j,3,:], 
                      color='k',linewidth=1,
                     density=1.5,arrowstyle='-|>')
        ax.text(0.02,0.85,dates1[j],transform=ax.transAxes,bbox=props,fontsize=24)
        
        j += 1
        

plt.tight_layout()
plt.show()
```
