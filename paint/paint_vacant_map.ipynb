```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt


# Set up a custom projection
projection = projection=ccrs.PlateCarree()

# Set up a figure
fig = plt.figure(dpi=150)
ax = fig.add_subplot(111, projection=projection)
ax.set_extent((30., 150., -10., 30.))

# Draw standard features
#ax.gridlines()
ax.coastlines()
ax.stock_img()
#ax.add_feature(cfeature.BORDERS)

ax.plot([30, 150], [0, 0], '--k', linewidth=0.8)
ax.plot([30, 150], [10, 10], '--k', linewidth=0.8)
ax.plot([80, 80], [-20, 50], '--k', linewidth=0.8)
ax.plot([90, 90], [-20, 50], '--k', linewidth=0.8)
ax.plot([100, 100], [-20, 50], '--k', linewidth=0.8)

#ax.set_title("Albers Equal Area Projection")


#plt.show()
plt.savefig('/home/sun/paint/map/south_asian.png', dpi=500)
```
